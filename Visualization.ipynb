## WordCloud

### Default plot

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba

text_from_file_with_apath = open('Business words.txt').read()

wordlist_after_jieba = jieba.cut(text_from_file_with_apath, cut_all = True)
wl_space_split = " ".join(wordlist_after_jieba)

my_wordcloud = WordCloud().generate(wl_space_split)

plt.imshow(my_wordcloud)
plt.axis("off")
plt.show()
```

### Modified plot

```python
from wordcloud import WordCloud
import PIL.Image as image
import numpy as np
import jieba
```

```python
def cut(text):
    word_list = jieba.cut(text,cut_all= True)
    result = " ".join(word_list)
    return result
```

```python
with open("Business words.txt") as fp:
    text = fp.read()
    text = cut(text)

mask = np.array(image.open("qq.jpg"))
wordcloud = WordCloud(mask = mask, background_color='#FFFFFF').generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
